```python
# import libray
from PIL import Image 
from pylab import *


##image import

im = Image.open('image/1.PNG')
print(im)
```

```python
crop = (75,100,180,150) 
im_crop = im.crop(crop) 
im_rs = im.resize((100,100)) 
im_rotate = im.rotate(180)

```

```python
print(im)
print(im_crop)
print(im_rs)
print(im_rotate)
imshow(im)
imshow(im_crop)
imshow(im_rs)
imshow(im_rotate)
```
